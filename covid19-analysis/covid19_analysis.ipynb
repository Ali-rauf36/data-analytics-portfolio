```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('owid-covid-data.csv')
df.head()
```

```python
df.shape

df.columns.tolist()

df.info()

df.describe()
```

```python
cols = [
    'continent', 'location', 'date', 
    'total_cases', 'new_cases',
    'total_deaths', 'new_deaths',
    'total_vaccinations', 'people_fully_vaccinated',
    'population'
]

df = df[cols]
df.shape
```

```python
df['date'] = pd.to_datetime(df['date'])
df['date'].dtype
```

```python
df.isnull().sum()
```

```python
# Rows with no continent are usually aggregates like 
# 'World', 'Asia', 'Europe' — not individual countries
df = df[df['continent'].notna()]
df.shape
```

```python
df['total_cases'] = df['total_cases'].fillna(0)
df['new_cases'] = df['new_cases'].fillna(0)
df['total_deaths'] = df['total_deaths'].fillna(0)
df['new_deaths'] = df['new_deaths'].fillna(0)
df['total_vaccinations'] = df['total_vaccinations'].fillna(0)
df['people_fully_vaccinated'] = df['people_fully_vaccinated'].fillna(0)

# Confirm no nulls remain
df.isnull().sum()
```

```python
# Get the latest date's data per country (most recent snapshot)
latest = df[df['date'] == df['date'].max()]

top10_cases = latest.nlargest(10, 'total_cases')[['location', 'total_cases']]

plt.figure(figsize=(10, 6))
sns.barplot(data=top10_cases, x='total_cases', y='location', palette='Reds_r')
plt.title('Top 10 Countries by Total COVID-19 Cases', fontsize=14, fontweight='bold')
plt.xlabel('Total Cases')
plt.ylabel('Country')
plt.tight_layout()
plt.show()
```

```python
print(df['total_cases'].dtype)
print(df['total_cases'].describe())
print(latest['total_cases'].value_counts().head())
```

```python
print(df['date'].max())
print(latest.shape)
print(latest['location'].head(10))
```

```python
# Get the last recorded row per country
latest = df.sort_values('date').groupby('location').last().reset_index()

print(latest.shape)
print(latest['total_cases'].describe())
```

```python
top10_cases = latest.nlargest(10, 'total_cases')[['location', 'total_cases']]

plt.figure(figsize=(10, 6))
sns.barplot(data=top10_cases, x='total_cases', y='location', palette='Reds_r')
plt.title('Top 10 Countries by Total COVID-19 Cases', fontsize=14, fontweight='bold')
plt.xlabel('Total Cases')
plt.ylabel('Country')
plt.tight_layout()
plt.show()
```

```python
top10_deaths = latest.nlargest(10, 'total_deaths')[['location', 'total_deaths']]

plt.figure(figsize=(10, 6))
sns.barplot(data=top10_deaths, x='total_deaths', y='location', palette='Greys_r')
plt.title('Top 10 Countries by Total COVID-19 Deaths', fontsize=14, fontweight='bold')
plt.xlabel('Total Deaths')
plt.ylabel('Country')
plt.tight_layout()
plt.show()
```

```python
global_trend = df.groupby('date')[['new_cases', 'new_deaths']].sum().reset_index()

fig, ax1 = plt.subplots(figsize=(14, 6))

ax1.plot(global_trend['date'], global_trend['new_cases'], 
         color='#3498db', linewidth=1.5, label='New Cases')
ax1.set_xlabel('Date')
ax1.set_ylabel('New Cases', color='#3498db')
ax1.tick_params(axis='y', labelcolor='#3498db')

ax2 = ax1.twinx()
ax2.plot(global_trend['date'], global_trend['new_deaths'], 
         color='#e74c3c', linewidth=1.5, label='New Deaths')
ax2.set_ylabel('New Deaths', color='#e74c3c')
ax2.tick_params(axis='y', labelcolor='#e74c3c')

plt.title('Global COVID-19 New Cases and Deaths Over Time', fontsize=14, fontweight='bold')
fig.legend(loc='upper left', bbox_to_anchor=(0.1, 0.95))
plt.tight_layout()
plt.show()
```

```python
top10_vax = latest.nlargest(10, 'people_fully_vaccinated')[
    ['location', 'people_fully_vaccinated', 'population']
].copy()

top10_vax['vax_pct'] = (top10_vax['people_fully_vaccinated'] / top10_vax['population']) * 100

plt.figure(figsize=(10, 6))
sns.barplot(data=top10_vax, x='vax_pct', y='location', palette='Greens_r')
plt.title('Top 10 Countries by Vaccination Rate (% of Population)', fontsize=14, fontweight='bold')
plt.xlabel('% Fully Vaccinated')
plt.ylabel('Country')
plt.axvline(x=70, color='red', linestyle='--', label='70% Herd Immunity Threshold')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
continent_stats = latest.groupby('continent').agg(
    total_cases=('total_cases', 'sum'),
    total_deaths=('total_deaths', 'sum')
).reset_index()

continent_stats['death_rate'] = round(
    (continent_stats['total_deaths'] / continent_stats['total_cases']) * 100, 2
)

plt.figure(figsize=(10, 6))
colors = ['#e74c3c' if x > 1 else '#2ecc71' for x in continent_stats['death_rate']]
sns.barplot(data=continent_stats, x='death_rate', y='continent', palette=colors)
plt.title('COVID-19 Death Rate by Continent (%)', fontsize=14, fontweight='bold')
plt.xlabel('Death Rate (%)')
plt.ylabel('Continent')
plt.axvline(x=1, color='black', linestyle='--', label='1% Threshold')
plt.legend()
plt.tight_layout()
plt.show()
```

## COVID-19 Global Data Analysis — Key Findings

**Dataset:** 333,420 records across 243 countries, January 2020 – October 2023

### Finding 1: Cases vs Deaths Tell Different Stories
Top 10 countries by cases and deaths overlap significantly but not completely.
China and South Korea rank high in cases but absent from deaths top 10 —
suggesting effective healthcare response or underreporting of deaths.
Mexico, Peru and Iran appear in deaths top 10 despite lower case counts —
indicating higher case fatality rates.

### Finding 2: Pandemic Timeline
- Early 2020: High death rates, artificially low case counts due to limited testing
- 2022: Massive Omicron wave caused the largest case spike of the entire pandemic
- 2023: Sharp decline as mandatory reporting ended globally

### Finding 3: Vaccination Coverage
Smaller nations (Uruguay, Greece, Hong Kong) achieved higher vaccination 
rates than large nations like the US and China — highlighting the difference 
between absolute numbers and population-adjusted rates.

### Finding 4: Continental Death Rates
Africa, North America and South America exceeded the 1% death rate threshold.
Africa's rate likely reflects case underreporting due to limited testing.
North and South America reflect overwhelmed healthcare systems during peak waves.
